# Identify columns that over 85% missing values


```python
def check_null_features(df, threshhold):
    nulls_percent_df = pd.DataFrame(df.isna().sum()/len(df)).reset_index()
    nulls_percent_df.columns = ['column_name', 'nulls_percentage']   
    columns_above_threshold = nulls_percent_df[nulls_percent_df['nulls_percentage']>threshhold]
    drop_columns_list = list(columns_above_threshold['column_name'])
    return drop_columns_list
```

```python
check_null_features(categorical, 0.85)
```

```python
droplist2 = check_null_features(categorical,0.85)
print(droplist2)
```

```python
for item in droplist2:
    droplist.append(item)
```

```python
droplist
```

# Remove those columns from the dataframe


```python
categorical = categorical.drop(droplist, axis = 1)
```

## Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

```python
categorical['GENDER'].unique()
```

```python
null_value_count = categorical["GENDER"].isna().sum()
null_value_count
```

```python
categorical['GENDER'] = categorical['GENDER'].fillna('F')
```

```python
def clean_gender_column(item):
    if item in ['F','M']:
        return item
    elif item in ['C', 'U', 'J', 'A']:
        return 'Other'
    else: 
        return item
```

```python
categorical['GENDER'] = categorical['GENDER'].apply(clean_gender_column)
```
